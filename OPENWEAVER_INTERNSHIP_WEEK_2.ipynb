```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Sample text data
text_data = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# Step 1: Tokenize the text data
# Use CountVectorizer for tokenization and word frequency calculation
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(text_data)

# Step 2: Calculate word frequencies
# Create a DataFrame to display word frequencies
word_freq_df = pd.DataFrame(X_count.toarray(), columns=count_vectorizer.get_feature_names_out())

# Step 3: Calculate TF-IDF vectors
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(text_data)

# Display the TF-IDF vectors and word frequencies
print("TF-IDF Vectors:")
print(X_tfidf.toarray())

print("\nWord Frequencies:")
print(word_freq_df)
```

```python
import spacy
from spacy.matcher import PhraseMatcher

# Load the SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Input text with multiple sentences
input_text = """
SpaCy is a popular NLP library.
It provides tools for natural language processing.
The PhraseMatcher in SpaCy is useful for finding specific phrases in text.
I love using SpaCy for text analysis.
"""

# Tokenize the input text
doc = nlp(input_text)

# Define a list of keywords or phrases to match
keywords = ["SpaCy", "PhraseMatcher", "text analysis"]

# Create a PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)

# Add the keywords as patterns
patterns = [nlp(keyword) for keyword in keywords]
matcher.add("Keywords", patterns)

# Find sentences that match the keywords
matches = matcher(doc)

# Extract and print matching sentences
matching_sentences = [doc[start:end].sent for match_id, start, end in matches]
for sentence in matching_sentences:
    print(sentence.text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data
text_data = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
tfidf_matrix = tfidf_vectorizer.fit_transform(text_data)

# Get the feature names (words) from the vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()

# Create a dictionary to store word frequencies
word_frequencies = {}

# Loop through each word and its frequency in the TF-IDF matrix
for i, word in enumerate(feature_names):
    frequency = tfidf_matrix.sum(axis=0).A1[i]
    word_frequencies[word] = frequency

# Sort the word frequencies by frequency in descending order
sorted_word_frequencies = sorted(word_frequencies.items(), key=lambda x: x[1], reverse=True)

# Print the word frequencies
for word, frequency in sorted_word_frequencies:
    print(f"{word}: {frequency}")
```

```python
import spacy

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample fake and real news articles (replace with your dataset)
fake_news_article = "According to a recent report, aliens have landed in New York City."
real_news_article = "The stock market reached a new record high today."

# Process the fake news article
fake_doc = nlp(fake_news_article)

# Process the real news article
real_doc = nlp(real_news_article)

# Extract entities from the fake news article
fake_entities = [(ent.text, ent.label_) for ent in fake_doc.ents]

# Extract entities from the real news article
real_entities = [(ent.text, ent.label_) for ent in real_doc.ents]

# Compare extracted entities
print("Entities in Fake News Article:")
for entity, label in fake_entities:
    print(f"{entity}: {label}")

print("\nEntities in Real News Article:")
for entity, label in real_entities:
    print(f"{entity}: {label}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample fake and real news articles (replace with your dataset)
fake_news_articles = [
    "According to a recent report, aliens have landed in New York City.",
    "Scientists discover a new species of dinosaur in South America.",
    # Add more fake news articles here
]

real_news_articles = [
    "The stock market reached a new record high today.",
    "World leaders gather for climate change summit in Paris.",
    # Add more real news articles here
]

# Combine all news articles into a single list
all_news_articles = fake_news_articles + real_news_articles

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
tfidf_matrix = tfidf_vectorizer.fit_transform(all_news_articles)

# Get the TF-IDF feature names (words)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Create a dictionary to store TF-IDF vectors for fake and real articles
tfidf_vectors = {
    'Fake News Articles': tfidf_matrix[:len(fake_news_articles)],
    'Real News Articles': tfidf_matrix[len(fake_news_articles):]
}

# Compare TF-IDF vectors
for category, tfidf_vector in tfidf_vectors.items():
    print(f"Category: {category}")
    print("Top TF-IDF terms:")
    top_terms_indices = tfidf_vector.sum(axis=0).argsort()[0, ::-1][:10]
    top_terms = [feature_names[i] for i in top_terms_indices]
    print(top_terms)
    print("\n")
```
