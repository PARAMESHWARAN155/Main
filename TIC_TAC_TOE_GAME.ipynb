```python
# Tic Tac Toe Game

# Initialize the board
board = [" " for _ in range(9)]

# Function to print the board
def print_board(board):
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("-" * 9)
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("-" * 9)
    print(f"{board[6]} | {board[7]} | {board[8]}")

# Function to check for a win
def check_win(board, player):
    # Check rows, columns, and diagonals for a win
    for i in range(3):
        if (
            (board[i] == player and board[i + 3] == player and board[i + 6] == player) or
            (board[i * 3] == player and board[i * 3 + 1] == player and board[i * 3 + 2] == player)
        ):
            return True
    if (
        (board[0] == player and board[4] == player and board[8] == player) or
        (board[2] == player and board[4] == player and board[6] == player)
    ):
        return True
    return False

# Function to check if the board is full
def check_full(board):
    return " " not in board

# Main game loop
current_player = "X"
winner = None
while not check_full(board) and winner is None:
    print_board(board)
    print(f"Player {current_player}'s turn:")
    move = int(input("Enter your move (1-9): ")) - 1

    if move >= 0 and move < 9 and board[move] == " ":
        board[move] = current_player
        if check_win(board, current_player):
            winner = current_player
        else:
            current_player = "O" if current_player == "X" else "X"
    else:
        print("Invalid move. Try again.")

print_board(board)

if winner:
    print(f"Player {winner} wins!")
else:
    print("It's a draw!")
```
