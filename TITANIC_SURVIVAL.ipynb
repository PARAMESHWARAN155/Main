```python


# Calculate correlations
correlation_matrix = df.corr()
data=[['Survival','Age','Sex','Embarked']]

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
